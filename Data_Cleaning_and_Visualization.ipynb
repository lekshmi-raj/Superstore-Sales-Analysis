# step 2
# top 10 customer according to there recency,frequency and revenue
# also find the top 10 customers revenue percentage

```python
# RFM 
# R : Recency
# F : Frequency
# M : Monitoring

# grouping
rfm = df.groupby("Customer ID").agg({
    "Order Date" : lambda x:(df["Order Date"].max() - x.max()).days, # recency
    "Order ID" : "count", # frequency
    "Sales" : "sum" # monitoring
}).round(2).reset_index()

# giving columns
rfm.columns = ["Customer ID","Recency","Frequency","Revenue"]

# create groups
rfm["R_Score"] = pd.qcut(rfm["Recency"],4,labels=[4,3,2,1])
rfm["F_Score"] = pd.qcut(rfm["Frequency"],4,labels=[1,2,3,4])
rfm["M_Score"] = pd.qcut(rfm["Revenue"],4,labels=[1,2,3,4])
rfm["RFM_Score"] = rfm["R_Score"].astype(str) + rfm["F_Score"].astype(str) + rfm["M_Score"].astype(str)

# finding top 10 customers
top_customers = rfm[rfm["RFM_Score"].str.startswith("444")].head(10)
plt.figure(figsize=(12,6))
sns.barplot(data=top_customers,x="Revenue",y="Customer ID",color="green",order=top_customers.sort_values("Revenue",ascending=True)["Customer ID"])
plt.title("Top 10 Customers",fontweight="bold",fontsize=15)
plt.tight_layout()
plt.savefig("top_10_customers.png",dpi=300,bbox_inches="tight")
plt.show()

# finding top 10 cutomers revenue percentage
top_rev_pct = top_customers["Revenue"].sum() / rfm["Revenue"].sum() * 100
print(f"{top_rev_pct:.1f}%")
print(rfm.head())
```

# step 3
# discount vs profit margin analysis

```python
# discount analysis by profit margin
plt.figure(figsize=(10,6))
sns.scatterplot(data=df.sample(5000),x="Discount",y="Profit Margin",hue="Category",alpha=0.6)
plt.title("Discount Analysis Based on Profit Margin",fontsize=20,fontweight="bold")
plt.axhline(y=0,color="red",linestyle="--",label="Break-even")
plt.legend()
plt.savefig("Discount_Analysis.png",dpi=300,bbox_inches="tight")
plt.show()

# worst performing sub category

subcat_profit = df.groupby(["Category","Sub-Category"])[["Sales","Profit"]].sum().reset_index()
subcat_profit["Margin"] = subcat_profit["Profit"] / subcat_profit["Sales"] * 100
worst = subcat_profit.nsmallest(5,"Margin")
print(worst[["Sub-Category","Margin"]])
```

# step 4
# monthly sales analysis

```python
monthly = df.groupby(["Year","Month"])[["Sales","Profit"]].sum().reset_index()

plt.figure(figsize=(16,6))
sns.lineplot(data=monthly,x="Month",y="Sales",hue="Year",marker="o")
plt.title("Monthly Sales Analysis",fontsize=20,fontweight="bold")
plt.savefig("Monthly_Sales_Analysis.png",dpi=300,bbox_inches="tight")
plt.show()

print("Q1 VS Q4 Sales Analysis")
Q4 = monthly[(monthly['Month'].isin(["October","November","December"]))]['Sales'].sum()
Q3 = monthly[(monthly["Month"].isin(["July","August","September"]))]["Sales"].sum()
Q2 = monthly[(monthly["Month"].isin(["April","May","June"]))]["Sales"].sum()
Q1 = monthly[(monthly['Month'].isin(["January","February","March"]))]['Sales'].sum()
print(f"Q1 sales is :${Q1:,.0f}")
print(f"Q2 sales :${Q2:,.0f}")
print(f"Q3 sales :${Q3:,.0f}")
print(f"Q4 sales is :${Q4:,.0f}")

print(f"Q3 is {((Q4 / Q3) * 100):.0f}% higher than Q4")
```
